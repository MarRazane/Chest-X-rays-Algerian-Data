```python
import matplotlib.pyplot as plt
```

```python

value_counts = data['Finding Labels'].value_counts()
plt.bar(value_counts.index, value_counts.values)
plt.xlabel('Unique Values')
plt.ylabel('Number of images')

# Adjust tick label properties
plt.xticks(rotation=45, ha='right', fontsize=10)  
plt.yticks(fontsize=10)
plt.show()

```

```python
import seaborn as sns
from PIL import Image
import os

# Set the seaborn style to mimic Tableau
sns.set(style="whitegrid")

# Plotting the images
fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(16, 12), dpi=100)

image_directory = 'data dz'  # Directory containing the images

for i, label in enumerate(value_counts.index):
    # Selecting images with the current label
    images = data[data['Finding Labels'] == label]['Image Index']

  # Displaying the first four images in the current label
    for j in range(min(4, len(images))):
        image_path = os.path.join(image_directory, images.iloc[j])
        image = Image.open(image_path)
        axs[i % 3, j].imshow(image)
        axs[i % 3, j].set_title(label)
        axs[i % 3, j].axis('off')

# Hide any empty subplots
if len(value_counts) < 12:
    for i in range(len(value_counts), 12):
        axs[i % 3, i % 4].axis('off')

plt.tight_layout()
plt.show()
```
